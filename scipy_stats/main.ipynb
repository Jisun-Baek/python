```python
import numpy as np
a = np.random.normal(loc=3.0, size=(2, 3))
b = np.random.normal(size=(3, 2)) 
print(a)
print(type(a))
# print(a.shape) # 변수 - attribute
# print(a.dtype) # 변수 - attribute
# print(a.T)     # 변수 - attribute
print(a.argmax(axis=1))  # 괄호있으면 - 함수 - method
print(a.reshape((1,-1))) # 괄호있으면 - 함수 - method
print(a.dot(b))          # 괄호있으면 - 함수 - method
print(dir(a))
```

```python
import scipy.stats as stats
a = stats.norm(loc=3.0)
samples = a.rvs(size=(2, 3), random_state=1)
print(samples)
print(type(samples))
print(samples.shape)
print(samples.dtype)
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
mu = 3.0
a = stats.norm(loc=mu)
x = np.linspace(mu-3, mu+3, 100)
y = a.pdf(x)
y2 = a.cdf(x)
plt.plot(x, y, label='pdf')
plt.plot(x, y2, label='cdf')
plt.legend(loc='lower left')
plt.show()
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
la = 3.0 
a = stats.expon(scale=1/la)
x = np.linspace(0., 3, 100)
y = a.pdf(x)
y2 = a.cdf(x)
plt.plot(x, y, label='pdf')
plt.plot(x, y2, label='cdf')
plt.legend(loc='lower left')
plt.show()
```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
mu = 3.0 
a = stats.poisson(mu)
x = np.linspace(0, 10, 11).astype(np.int64)
y = a.pmf(x)
y2 = a.cdf(x)
plt.bar(x, y2, label='cdf', alpha=0.5)
plt.bar(x, y, label='pmf', alpha=0.5)
plt.legend()
plt.show()
```
