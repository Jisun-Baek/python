```python
import numpy as np
import matplotlib.pyplot as plt 
import time

record_time = np.empty((1000000,))
temp = []
for i in range(1000000):
    tic = time.time()
    temp.append(i)
    toc = time.time()
    record_time[i] = toc-tic
    
plt.plot(record_time)
plt.savefig('img/record_time.png')
plt.show()
plt.close('all')
```
